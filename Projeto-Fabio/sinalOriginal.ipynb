```python
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
import sounddevice as sd
```

```python
sinal, taxaAmostragem = sf.read('audioMario.wav')
print(taxaAmostragem)
```

Sinal original:

```python
t = np.linspace(0, len(sinal)/taxaAmostragem, len(sinal))
plt.plot(t, sinal)
plt.title("Áudio no tempo")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
transformadaFourierOriginal = fft(sinal)
frequencias = fftfreq(len(transformadaFourierOriginal), 1/taxaAmostragem)
```

```python
plt.plot(frequencias[:len(transformadaFourierOriginal)//12], np.abs(transformadaFourierOriginal[:len(transformadaFourierOriginal)//12]))
plt.title("Espectro de Frequência")
plt.xlabel("Frequência (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.xlim(0, 4000) 
plt.ylim(0, 2500) 
plt.show()
```

```python
sd.play(sinal, taxaAmostragem)
```
